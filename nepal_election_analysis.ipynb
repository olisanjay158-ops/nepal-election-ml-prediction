```python
cands = sorted(set(w2022_n["Constituency"]))
auto_map = {}

for m in missing_2022:
    best = None
    best_score = 0.0
    for c in cands:
        s = difflib.SequenceMatcher(None, m.lower(), c.lower()).ratio()
        if s > best_score:
            best_score = s
            best = c
    if best_score >= 0.80:
        auto_map[best] = m

auto_map
```
